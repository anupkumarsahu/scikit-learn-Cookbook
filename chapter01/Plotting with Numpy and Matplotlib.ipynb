```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot a straight line by placing two NumPy arrays of the same length:

```python
plt.plot(np.arange(10), np.arange(10))
```

Plot an exponential:

```python
plt.plot(np.arange(10), np.exp(np.arange(10)))
```

Place the two graphs side by side:

```python
plt.figure()
plt.subplot(121)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(122)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

top to bottom:

```python
plt.figure()
plt.subplot(211)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(212)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

Plot in a 2 x 2 grid traversing in reading order from one to four:

```python
plt.figure()
plt.subplot(221)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(222)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(223)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(224)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
target = iris.target

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.scatter(data[:,0], data[:,1], c=target)

plt.subplot(122)
plt.scatter(data[:,2], data[:,3], c=target)
```
